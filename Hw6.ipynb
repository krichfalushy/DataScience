```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_1 = "/content/drive/MyDrive/Colab Notebooks/data_1/data_2d.csv"
data_2d = pd.read_csv(data_1)
data_2d.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
inertia_2d = []
for i in range(1, 16):
  cluster = KMeans(n_clusters=i, random_state=0, n_init="auto").fit(data_2d)
  inertia_2d.append(cluster.inertia_)

plt.grid()
plt.plot([i for i in range(1, 16)], inertia_2d, marker=".")
plt.show()
```

```python
data_2 = "/content/drive/MyDrive/Colab Notebooks/data_1/mnist.csv"
data_mnist = pd.read_csv(data_2)
data_mnist
```

```python
inertia_mnist = []
for i in range(1, 16):
  cluster = KMeans(n_clusters=i, random_state=0, n_init="auto").fit(data_mnist)
  inertia_mnist.append(cluster.inertia_)

plt.grid()
plt.plot([i for i in range(1, 16)], inertia_mnist, marker=".")
plt.show()
```

```python
model_2d = KMeans(n_clusters=2, random_state=0, n_init="auto").fit(data_2d)
# model_mnist = KMeans(n_clusters=4, random_state=0, n_init="auto").fit(data_mnist)
```

```python
label_1 = model_2d.labels_
plt.figure(figsize=[10, 7])
plt.grid()
plt.scatter(data_2d["-7.687164597386728637e-01"], data_2d["4.608603078297135447e-01"], c=label_1)
plt.title("Model_2d Visualisation")
plt.show()
```

```python
pca = PCA(n_components=2).fit_transform(data_mnist)
pca = pd.DataFrame(pca)
model_mnist = KMeans(n_clusters=5, random_state=0, n_init = "auto").fit(pca)
plt.figure(figsize=[10, 7])
plt.grid()
plt.title("Model_Mnist Visualisation")
plt.scatter(pca[0], pca[1], c=model_mnist.labels_)
plt.show()
```
