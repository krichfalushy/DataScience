```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tables = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8", match="Регіон", thousands=".", decimal=",")
table = tables[0]
print(table)
```

```python
print(table.head())
```

```python
print(table.shape)
rows, columns = table.shape
print(f"К-сть рядків: {rows}\nК-сть стовпців: {columns}")
```

```python
table.replace("—", "NaN")
```

```python
repl_table = table.replace(to_replace="—", value=np.NaN)
print(repl_table)
print(f"\nТипи стовпців: \n{repl_table.dtypes}")
```

```python
num_table = repl_table.apply(pd.to_numeric, errors="ignore")
print(num_table)
num_table.dtypes
```

```python
null_table = repl_table.isnull().sum(axis=1)
print(null_table)
```

```python
del_table = repl_table.drop([27], axis=0)
print(del_table)
```

```python
fillna_table = num_table.fillna(num_table.mean())
print(fillna_table)
```

```python
table_2019 = fillna_table["2019"]
region_table = fillna_table["Регіон"]
indexes = []
regions = []
values = []
ind = 0
while ind <= 27:
  for i in table_2019:
    if float(i) > float(table_2019[27]):
      indexes.append(ind)
      values.append(i)
    ind += 1

ind = 0
for i in region_table:
  if ind in indexes:
    regions.append(i)
  ind += 1

result_2019 = pd.DataFrame({"Регіон": regions,
                            "2019": values}, index=indexes)

# print(table_2019, indexes)
# print(region_table)
# print(fillna_table.round())
print(result_2019)

# fillna_table[fillna_table["2019"] > fillna_table["2019"].mean()][["Регіон", "2019"]]
```
